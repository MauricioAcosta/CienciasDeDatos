```python
#Carga de dataset
docentes=pd.read_csv('Est_mulos_Universitarios_-_Universidad_de_la_Amazonia_II_-2014.csv')
print(docentes)
```

```python
docentes_nan = docentes.isnull()
print(docentes_nan)
# Efecto: retorna todos los indices con un valor booleano los valores que no estan vacios, false para campo lleno, y true para campos vacios
```

```python
docentes_sin_nan1 = docentes.dropna(axis=0, how='any')
print(docentes_sin_nan1)
# Retorna los campos que tengan algun valor y elimina los que no tienen asignado ningun valor, donde axis=0 son los indices y 
# how='any' elimina la fila
```

```python
docentes_sin_nan2 = docentes.dropna(axis=0, how='all')
print(docentes_sin_nan2)
# Retorna todos los campos y los que no tienen asignado ningun valor les asigna por defecto NaN, donde axis=0 son los indices y how='all' son todos los registros
```

```python
docentes_sin_nan3 = docentes.dropna(axis=1, how='any')
print(docentes_sin_nan3)
# limpia (axis=1) la columna para cualquier valor que este vacio (how='any') 
```

```python
docentes_nuevo = docentes.fillna(value=3)
print(docentes_nuevo)
# completa los valores vacios, con el valor asignado, en este caso value=3
```

```python
docentes_nuevo = docentes.fillna(docentes.mean())
print(docentes_nuevo)
# Para cualquier valor vacío, completa con el valor del promedio
```

```python
docentes_nuevo = docentes.fillna(method='bfill', limit=1)
print(docentes_nuevo)
# Completa los campos vacios, con el valor por debajo del campo vacío, y solo remplazara el primer campo que encuentre vacio por el método limit=1
```

```python
# Efecto:
# Remplaza los valores donde la condición es falsa por NaN y muestra todos los valores que son mayores a 5 en el dataframe
docentes.where(docentes > 5)
```

# ¿Cuál es la principal diferencia entre where y mask?

```python
# Efecto:
# Remplaza los valores donde la condición es verdadera y muestra todos los valores que son menores a 5 en el dataframe
docentes.mask(docentes > 5)
```

```python
# La diferencia entre where y mask, es que mientras que where muestra los valores mayores a la condición, mask muestra los valores menores y que cumplen la condición, es decir si 3<5 == true.
```

```python
# Selecciona los datos en el dataframe que están en el arreglo que se ha pasado 
# como parametro de la función, colocando en true los valores que sean igual a los del arreglo, y false los que no.
docentes.isin([2,3,4,5,7])
```
